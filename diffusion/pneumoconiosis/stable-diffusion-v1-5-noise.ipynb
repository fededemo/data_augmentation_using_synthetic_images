```python
import requests
import torch
from PIL import Image
from io import BytesIO
from os import makedirs, listdir

from diffusers import StableDiffusionImg2ImgPipeline

from os.path import exists, join
import torch
```

```python
if exists("/content/drive/MyDrive/ORT/Tésis Máster/Codes/datasets/generated_datasets/pneumoconiosis/diffusion/StableDiffusionV1"):
  WORK_DIR = "/content/drive/MyDrive/ORT/Tésis Máster/Codes/datasets/generated_datasets/pneumoconiosis/diffusion/StableDiffusionV1"
  DATA_DIR = '/content/drive/MyDrive/ORT/Tésis Máster/Codes/datasets/transformed_datasets/pneumoconiosis/pneumoconiosis_resized/train/1'
else:
  WORK_DIR = "/content/drive/MyDrive/ORT/Tesis/Codes/datasets/generated_datasets/pneumoconiosis/diffusion/StableDiffusionV1"
  DATA_DIR = '/content/drive/MyDrive/ORT/Tesis/Codes/datasets/transformed_datasets/pneumoconiosis/pneumoconiosis_resized/train/1'

print(WORK_DIR)
print(DATA_DIR)
```

```python
DEVICE = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
DEVICE
```

```python
model_id_or_path = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(model_id_or_path, safety_checker=None, torch_dtype=torch.float16)
pipe = pipe.to(DEVICE)
```

```python
g_cuda = torch.Generator(device=DEVICE)
g_cuda.manual_seed(1337)
```

```python
NOISE_ADDED = 0.05
INFERENCE_STEPS = 50
```

```python
NOISE_STR = f'noise_{NOISE_ADDED}_{INFERENCE_STEPS}'
if not exists(join(WORK_DIR, NOISE_STR)):
    makedirs(join(WORK_DIR, NOISE_STR))
```

```python
for image in listdir(DATA_DIR):
    init_image = Image.open(join(DATA_DIR,image)).convert("RGB")
    prompt = "a chest-xray image"
    images = pipe(prompt=prompt, image=init_image, strength=NOISE_ADDED, guidance_scale=7.5, num_inference_steps=INFERENCE_STEPS, generator=g_cuda).images
    images[0].save(join(WORK_DIR, NOISE_STR, f'noise_{NOISE_ADDED}_{INFERENCE_STEPS}_{image}'))
```
