```python
import torch
import transformers
from accelerate import Accelerator
from accelerate.logging import get_logger
from accelerate.utils import ProjectConfiguration, set_seed
from huggingface_hub import create_repo, model_info, upload_folder
from packaging import version
from PIL import Image
from torchvision import transforms
from transformers import AutoTokenizer, PretrainedConfig
from tqdm.auto import tqdm

import diffusers
from diffusers import (
    AutoencoderKL,
    DDPMScheduler,
    DiffusionPipeline,
    DPMSolverMultistepScheduler,
    StableDiffusionPipeline,
    UNet2DConditionModel,
)
```

```python
pipeline = DiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float32,
        safety_checker=None,
        revision=None,
    )

pipeline = pipeline.to("cuda")
pipeline.set_progress_bar_config(disable=True)

prompts = [
    ("pneumoconiosis", "image of pneumoconiosis xray"),
    ("glioma", "an image of a glioma-brain-tumor MRI"),
    ("retinopathy", "an image of a mild retinopathy")
    ]

imgs_count = 3
for cls, prompt in prompts:
    print(f"Generating for {cls}")
    for ix, img in tqdm(enumerate(pipeline(prompt, num_images_per_prompt=imgs_count).images), total=imgs_count):
        img.save(f"{cls}_{ix}.png")
```
